## Video of the EDI globaly over the years
To start out our presentation we decided to engage our class by generating a video. This also worked quite well to show the general evolution of democracy for the whole world. The main problem with creating the individual frames for the video was that the geoBoundaries dataset [Runfola, D. et al. (2020)], which we used for plotting countries for each year, contains the actual names of countries as they were in the corresponding year. The edi dataset in comparison just uses the current country names for every year. This lead to us leading to manually rename many of the countries in the geoBoundaries dataset to be able to join the edi data.<br>
To actually plot the data we used geopandas to read the polygon information for each country from the geoBoundaries dataset, joined the edi data on the country name and year ant then colored the countries based on their edi value. We did this for every year in our dataset and then used open-cv to generate a video. For the framerate we chose 3 as this allows the viewer to still see each frame without making the final video too long and thus boring.<br>
One issue with the edi dataset is that very vew countries were analysed for the early years. This leads to a lot of countries being displayed as grey for most of the video.

## The average evolution of democracy

The video above already did an adequate job of showing how democracy evolved but to further consolidate that the world has gotten more democratic in general we decided that we need some plot which shows the average edi for different years. To also show it's spread we decided on using boxplots. As we have more than 200 years to cover we decided to not generate a plot for every year but instead only plotted every tenth year up until 2000 and then every single year because recent history interested us more. We could also have done a lineplot with the variance or standard deviation depicted as lines around the mean.<br>
The plot does a good job of showing that the electoral democracy index has increased quite a lot since 1790 even though there are still many countries with very low values as indicated by the lower whiskers which are close to zero for all years.<br>
This plot looks a little boring due to bein just black and white and we thought about spicing it up by adding colours, but this would not have helped supplementing the information shown which is why we decided to keep the plot as is.

## Political Regimes
We mostly used the electoral democracy index for our plots, but the regimes dataset offered some more interesting insights in what ways democracy has evolved. The classifications in this dataset are Based on Lhrmann et al. (2018) and the estimates by V-Dems experts. Many countries are classified as one of the following regimes for each year since 1970:
- Closed autocracy: Power is concentrated in the hands of one person or small group
- Elected autocracy: Elected leader governs in an authortarian manner
- Elected democracy: Elected leaders through free and fair elections. Government operates on principles of majority rule
- Liberal democracy: Free and fair elections with the protection of individual rights and liberties, the rule of law and separation of powers
To show how the distribution of regimes has changed we decided to stack the number of countries for each type of regime for every year.<br>
The large increase in countries at 1900 is caused by the dataset containing a lot more analyzed countries after 1900 compared to before.
